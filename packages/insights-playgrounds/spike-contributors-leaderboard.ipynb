# Metric Contributions 
a contribution is any activity set by as isContribution = true

```python
%%sql
select 
members."attributes" -> 'avatarUrl' ->>'github' "logo_url",
activities."username",
count(*) activities_count
from activities
left join members on members.id = activities."memberId"
where activities."tenantId"='ccff5355-cf54-40a1-9a2e-8e4a447ae73a'
and "isContribution" = true
and timestamp > current_date - interval '7 days'
group by 1,2
order by activities_count desc
```
