### how to count contributions for a project or repository ? 


```python
%%sql
select * from activities limit 5
```

```python
%%sql
select 
"tenantId", "isContribution",  type, attributes, url
from activities 
where "tenantId" = 'ccff5355-cf54-40a1-9a2e-8e4a447ae73a'
and url like '%pull/3929%'
order by timestamp desc
limit 50

```

```python
%%sql
select 
"tenantId", "isContribution",  type, attributes, url,timestamp
from activities 
where "tenantId" = 'ccff5355-cf54-40a1-9a2e-8e4a447ae73a'
and "isContribution" = true
and timestamp > current_date - interval '7 day'
order by timestamp desc
limit 50
```

- issues-opened
- pull_request-merged
- pull_request-review-thread-comment
- pull_request-reviewed
- pull_request-closed

```python
%%sql
select 
distinct type
from activities 
where "tenantId" = 'ccff5355-cf54-40a1-9a2e-8e4a447ae73a'
and "isContribution" = true
and timestamp > current_date - interval '1 year'
limit 50

```

## Let's create the queries for the expected values 

1. Total contributions in a project in a selected date range
2. Contribution series

```python
%%sql
select 
count(*)
from activities 
where "tenantId" = 'ccff5355-cf54-40a1-9a2e-8e4a447ae73a'
and "isContribution" = true
and timestamp > current_date - interval '7 day'
```

```python
%%sql
select 
date_trunc('day', timestamp) as day, count(*)
from activities 
where "tenantId" = 'ccff5355-cf54-40a1-9a2e-8e4a447ae73a'
and "isContribution" = true
and timestamp > current_date - interval '7 day'
group by 1
order by 1 desc
```
